```python
plt.imshow(img)
plt.axis('off')  # 关闭坐标轴
plt.show()
```

```python
# read camera
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    cv2.imshow('frame', frame)
    #show r g b channel
    r, g, b = cv2.split(frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
